Downloading tweets and saving them into tweet_json_list

```python
# downloading with the Twitter API
count = 0
tweet_json_list = []
#fails_dict = {}

start_time = time.time()

for tweet_id in tweet_ids:
    count += 1
    try:
        # getting the tweet's JSON data and appending it to the tweet JSON list
        tweet = api.get_status(tweet_id, tweet_mode = 'extended')
        tweet_json_list.append(tweet._json)
    except tweepy.TweepError as err:
        pass
print("Number of loops: "+str(count))   

end_time = time.time()        
loop_time = end_time - start_time

print("\nLoop time (HH:MM:SS):  " + time.strftime("%H:%M:%S", time.gmtime(loop_time)) + "\n")

```

Function to write tweets(from:https://galeascience.wordpress.com/2016/03/18/collecting-twitter-data-with-python/)

```python
def write_tweets(tweets, filename):
    ''' Function that appends tweets to a file. '''
 
    with open(filename, 'a') as f:
        for tweet in tweets:
            json.dump(tweet, f)
            f.write('\n')
```

Writing actual tweets

```python
write_tweets(tweet_json_list, 'tweet_json.txt')
```

You can assess data for:

Quality: issues with content. Low quality data is also known as dirty data.

Tidiness: issues with structure that prevent easy analysis. Untidy data is also known as messy data.

Tidy data requirements:
Each variable forms a column.
Each observation forms a row.
Each type of observational unit forms a table.
...using two types of assessment:



### The four main data quality dimensions are:

#### Completeness: 
do we have all of the records that we should? Do we have missing records or not? Are there specific rows, columns, or cells missing?
#### Validity: 
we have the records, but they're not valid, i.e., they don't conform to a defined schema. A schema is a defined set of rules for data. These rules can be real-world constraints (e.g. negative height is impossible) and table-specific constraints (e.g. unique key constraints in tables).
#### Accuracy: 
inaccurate data is wrong data that is valid. It adheres to the defined schema, but it is still incorrect. 
Example: a patient's weight that is 5 lbs too heavy because the scale was faulty.
#### Consistency: 
inconsistent data is both valid and accurate, but there are multiple correct ways of referring to the same thing. Consistency, i.e., a standard format, in columns that represent the same data across tables and/or within tables is desired.
Regarding the other data quality research mentioned in the video, the additional dimensions are super specific cases of these four dimensions listed above. Example: currency, defined as follows: the degree to which data is current with the world that it models. Currency can measure how up-to-date data is. Currency is a specific case of accuracy data in the sense that out-of-date data is (usually) valid but wrong. In other words, our definition of accuracy can include currency.

Function from https://medium.com/@jorlugaqui/how-to-strip-html-tags-from-a-string-in-python-7cb81a2bbf44.

```python
def remove_html_tags(text):
    """Remove html tags from a string"""
    import re
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
```

```python
archive_clean['source'] = archive_clean['source'].apply(remove_html_tags)
```

#### 6. The not a dog breed values in columns p1, p2, p3('hotdog', 'seashore', 'hummingbird', etc.) in predictions dataframe should be handled;

##### Define

Clarifying values of dog breeds, removing odd values from the `p1, p2, p3` columns.

#### Code

Firstly, I download the dictionary of dog breeds from `https://www.kaggle.com/c/dog-breed-identification/data `.

```python
breeds = pd.read_csv('labels.csv')
```

```python
predictions_clean['p1'] = predictions_clean['p1'].str.lower()
predictions_clean['p2'] = predictions_clean['p2'].str.lower()
predictions_clean['p3'] = predictions_clean['p3'].str.lower()
```

Replacing non-dog names in `p1, p2, p3` columns with `None`

```python
for index, row in predictions_clean.iterrows():
    if not(row['p1'] in breeds_list):
        predictions_clean.at[index, 'p1']= None
        predictions_clean.at[index, 'p1_dog']= False
    if not(row['p2'] in breeds_list):
        predictions_clean.at[index, 'p2_dog']= False
    if not(row['p3'] in breeds_list):
        predictions_clean.at[index, 'p3_dog']= False
```

#### 6. One probability instead of three

##### Define

Leaving max probability out of 3 possibilities.

##### Code

Creating max probability columns.

```python
p = []
p_conf = []
for index, row in predictions_clean.iterrows():
    d = {}
    if row['p1_dog']:
            d[row['p1']]=row['p1_conf']
    if row['p2_dog']:
        d[row['p2']]=row['p2_conf']
    if row['p3_dog']:
        d[row['p3']]=row['p3_conf']
    print(d)
    if d:
        m = max(d.values())
        p_conf.append(max(d.values()))
        for k, v in d.items():
            if v==m:
                p.append(k)
    else:
        p_conf.append(None)
        p.append(None)
    print(p_conf[index], p[index])
```

```python
predictions_clean['p']=p
predictions_clean['p_conf']=p_conf
```
